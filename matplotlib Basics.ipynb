## Examples

Creating the following plot

```python
res_a1 = 0.2*np.random.randn(1000)+0.4
res_b1 = 0.25*np.random.randn(1000)+0.4
res_a2 = 0.21*np.random.randn(1000)+0.3
res_b2 = 0.22*np.random.randn(1000)+0.3
```

Good habit to come up with complicated strings beforehand

```python
textstr1 = '\n'.join((
    r'$\sigma_a=%.4f$' % (np.std(res_a1)),
    r'$\sigma_b=%.4f$' % (np.std(res_b1))))

textstr2 = '\n'.join((
    r'$\sigma_a=%.4f$' % (np.std(res_a2)),
    r'$\sigma_b=%.4f$' % (np.std(res_b2))))

# .4f means till four decimal places
```

Making the plot

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3.5))
ax = axes[0]
ax.hist(res_a1, bins=30, density=True, histtype='step', label='Method a', color='red')
ax.hist(res_b1, bins=30, density=True, histtype='step', label='Method b', color='blue')
ax.legend(loc='upper right', fontsize=10, edgecolor='black')
ax.text(0.05, 0.81, textstr1, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'), size=12 )
ax.set_ylabel('Density')
ax.set_title('Trial 1')

ax = axes[1]
ax.hist(res_a2, bins=30, density=True, histtype='step', label='Method 1', color='green')
ax.hist(res_b2, bins=30, density=True, histtype='step', label='Method 2', color='brown')
ax.legend(loc='upper right', fontsize=10, edgecolor='black')
ax.text(0.05, 0.81, textstr2, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black'), size=12 )
ax.set_title('Trial 2')


fig.text(0.5, -0.04, r'$\Delta E$[joules]', ha='center', fontsize=20) #ha: horizontal alignment
plt.show()
```

## Stream Plots

For plotting vector fields

```python
w = 3
_ = np.linspace(-3,3,100)
X, Y = np.meshgrid(_,_)
U = -1 - X**2 + Y
V = 1 + X - Y**2
speed = np.sqrt(U**2 + V**2)
```

```python
fig, axes = plt.subplots(2,2,figsize=(7,7))
ax = axes[0][0]
ax.streamplot(X,Y,U,V) #ax.streamplot() requires the X,Y meshgrids for points on the surface, and U,V components/conditions
ax = axes[0][1]
ax.streamplot(X,Y,U,V, color=speed, cmap='plasma') #color varies with speed
ax = axes[1][0]
lw = 5*speed/speed.max() #defining line width proportional to speed
ax.streamplot(X,Y,U,V,linewidth=lw) #linewidth varies with speed
ax = axes[1][1]
seed_points = np.array ([[0,1],[1,0]]) #setting starting points
ax.streamplot(X, Y, U, V, color=U, linewidth=2,cmap='autumn', start_points=seed_points) #shows the line of flow from the starting points
```

# Animations

Animations require the time of your problem to frame the animation, then frames per second.

$\bullet$ Below is a `sin` wave moving at $3ms^{-1}$. Translate this into an animation.

```python
def f(x,t):
    return np.sin(x-3*t)

x = np.linspace(0, 10*np.pi, 1000)
#t = np.arange(0, 24, 1/60)
#X, T = np.meshgrid(x, t)
#F = f(X, T)
```

```python
plt.plot(x,f(x,0), label='t=0.0s')
plt.plot(x,f(x,0.5), label='t=0.5s')
plt.plot(x,f(x,1), label='t=1.0s')
plt.plot(x,f(x,1.5), label='t=1.5s')
plt.legend()
plt.show()
```

Create Animation

```python
#importing the animation libraries
from matplotlib import animation
from matplotlib.animation import PillowWriter
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,4))
ln1, = plt.plot([], [])
time_text = ax.text(0.65, 0.95, '', fontsize=15,
                    bbox=dict(facecolor='white', edgecolor='black'), 
                    transform=ax.transAxes)
ax.set_xlim(0,10*np.pi)
ax.set_ylim(-1.5, 1.5)

def animate(i):
    ln1.set_data(x, f(x, 1/50*i))
    time_text.set_text('t={:.2f}'.format(i/60))
    
ani = animation.FuncAnimation(fig, animate, frames=240, interval=50)
ani.save('images/ani.gif',writer='pillow',fps=50,dpi=100)
```

The animation function can be used in other ways too. One can create 3D rotating gifs of surfaces.

```python
_ = np.linspace(-1, 1, 100)
x, y = np.meshgrid(_,_)
z = x**2+x*y
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(x,y,z, cmap='coolwarm',
                       linewidth=0, antialiased=False)
ax.view_init(elev=10, azim=0)

def animate(i):
    ax.view_init(elev=10, azim=1*i) #every frame rotates 1 degree
    
ani = animation.FuncAnimation(fig, animate, frames=360, interval=50) #since every frame rotates by 1 degree, we require 360 frames to cover one full rotation.
ani.save('images/ani2.gif',writer='pillow',fps=30,dpi=100)
```
