```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook', 'grid'])
x = [0,5,10,15,25]
y = [0.090, 0.152, 0.301, 0.372, 0.423]
plt.figure(figsize=(8, 5), dpi=300)
plt.plot(x, y, 'o--', color='red', lw=0.7, ms=2.5, label='Abs. at 540 nm')
plt.xlabel(r'Sample Concentration ($\mu g mL^{-1}$)')
plt.ylabel("Average Absorbance")
plt.legend(loc='upper left', fontsize=11)
plt.savefig('foo.png', bbox_inches='tight')
plt.show()
```

```python
import numpy as np
plt.style.use(['science', 'notebook', 'grid'])
plt.figure(figsize=(8, 5), dpi=300)
plt.plot(x, y, color='red', lw=0.7, ms=2.5, label='Actual Data')
coef = np.polyfit(x,y,1)
poly1d_fn = np.poly1d(coef)
plt.plot(x,y, 'yo', x, poly1d_fn(x), '--k', lw=0.7, ms=2.5, label="Fitted Curve")
print (poly1d_fn)
plt.xlabel(r'Sample Concentration ($\mu g mL^{-1}$)')
plt.ylabel("Average Absorbance")
plt.legend(loc='upper left', fontsize=11)
plt.savefig('poo.png', bbox_inches='tight')
plt.show()
```
