## Tabulation Method

Suppose I have an equation, $$10^{x} + x - 4 = 0$$ How do I find the interval between which the root lies, knowing that it occurs in $[0,1]$

```python
import math

def func(x):
	return 10**x + x - 4

i=0
while(i<=1): 
    if func(i)*func(i+0.01) < 0 :#consider the interval length to be 0.01
        lower = i
        upper = i+0.01
        break
    i = i + 0.01
    
print("The interval in which the root lies is: [%f, %f]"%(lower,upper))
```

```python
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook', 'grid'])

import numpy as np

sample = np.linspace(0, 1, 100)
plt.plot(sample, func(sample))
plt.plot(sample, np.zeros(100))
```

## Bisection Method

Use tabulation method to find a sufficiently small initial interval $[a_0, b_0]$ containing the root $\alpha$ of the equation $f (x) = 0$ such that $f (a_0)$ and $f (b_0)$ are of opposite signs. Then proceed with Bisection Method.

```python
import math

def func(x):
	return 10**x + x - 4

i=0
while(i<=1): 
    if func(i)*func(i+0.01) < 0 :#consider the interval length to be 0.01
        lower = i
        upper = i+0.01
        break
    i = i + 0.01
    
print("The interval in which the root lies is: [%f, %f]"%(lower,upper))
```

```python
def bisection (a_0, b_0):
    if func(a_0)*func(b_0)>0:
        print("Invalid Interval!")
        return

    while ((b_0-a_0)>=0.000000000000000000000000001):
        mid = (a_0 + b_0)/2
    
        if func(a_0) * func(mid) < 0:
            b_0 = mid
        else:
            a_0 = mid
        
    return (a_0, b_0, mid)

x, y, z = bisection(lower, upper)

print("The root according to Bisection method is: %8f and, the error in the value is: %8f" %(z, abs(func(z)-0)))
```

## Newton - Raphson Method

```python
def dfdx(x):
    return math.log(10)*10**x + 1

def newRap(x):
    h = func(x)/dfdx(x)
    
    while abs(h) >= 0.0001:
        h = func(x)/dfdx(x)
         
        # x(n+1) = x(n) - f(x) / f'(x)
        x = x - h
     
    return x
    
newRap(0.53)
```
