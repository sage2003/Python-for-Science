```python
import numpy as np
import sympy as smp
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# as nothing worked so resorting to the last solution
def d_r(m,r,p_r,theta,p_theta):
    return(p_r/m)

def d_theta(m,r,p_r,theta,p_theta):
    return(p_theta/(m*(r+a)**2))

def d_p_r(m,r,p_r,theta,p_theta):
    return(-m*g-k*r+m*g*np.cos(theta)+2*p_theta**2/(m*(r+a)**3))

def d_p_theta(m,r,p_r,theta,p_theta):
    return(-m*g*(r+a)*np.sin(theta))
```

```python
def dSdt(S,t):
    r, p_r, theta, p_theta=S
    return[
        d_r(m,r,p_r,theta,p_theta),
        d_p_r(m,r,p_r,theta,p_theta),
        d_theta(m,r,p_r,theta,p_theta),
        d_p_theta(m,r,p_r,theta,p_theta),
    ]
```

```python
t=np.linspace(0,50,1000)

ans=odeint(dSdt, y0=[xi,pxi,thetai,pthetai],t=t)
```

```python
m=1
g=10
k=10
a=5
rand_arr=np.random.uniform(1.01, 0.99, 800)
```

```python
xi=rand_arr[0:200]
pxi=rand_arr[200:400]
thetai=rand_arr[400:600]
pthetai=rand_arr[600:800]
```

```python
for i in range(200):
    ans=odeint(dSdt, y0=[xi[i],pxi[i],thetai[i],pthetai[i]],t=t)
    np.savetxt(f"run_{i}.csv",ans,delimiter=',')
```

```python
plt.plot(ans.T[0])
```

```python
# plotting the results
data=np.zeros((200,1000,4))
for i in range(200):
    ld_data=np.loadtxt(f"run_{i}.csv",delimiter=',')
    data[i,:,:]=ld_data
```

```python
# data[run,time,var]
plt.plot(data[100,::,0])
```

```python
i=10
plt.plot(data[::,i,1])
```

```python
p1_xmax=max(data[::,::,0].flatten())
p1_xmin=min(data[::,::,0].flatten())
p1_ymax=max(data[::,::,1].flatten())
p1_ymin=min(data[::,::,1].flatten())

p2_xmax=max(data[::,::,2].flatten())
p2_xmin=min(data[::,::,2].flatten())
p2_ymax=max(data[::,::,3].flatten())
p2_ymin=min(data[::,::,3].flatten())

for frame in range(0,100):
    fig=plt.figure(figsize=(15,15),facecolor='white')

    gs=fig.add_gridspec(2,1)

    ax1=fig.add_subplot(gs[0])
    ax2=fig.add_subplot(gs[1])

    ax1.scatter(data[::,frame,0],data[::,frame,1],alpha=0.7,color='red',edgecolors='black',s=25)
    ax1.set_xlim([p1_xmin,p1_xmax])
    ax1.set_ylim([p1_ymin,p1_ymax])
    ax1.set_xlabel(r"$x$",fontsize=14)
    ax1.set_ylabel(r"$P_x$",fontsize=14)

    ax2.scatter(data[::,frame,2],data[::,frame,3],alpha=0.7,color='red',edgecolors='black',s=25)
    ax2.set_xlim([p2_xmin,p2_xmax])
    ax2.set_ylim([p2_ymin,p2_ymax])
    ax2.set_xlabel(r"$\theta$",fontsize=14)
    ax2.set_ylabel(r"$P_{\theta}$",fontsize=14)


    plt.suptitle("Spring Pendulum",fontsize=16)
    plt.tight_layout()

    plt.savefig("{:05d}.png".format(int(frame)),dpi=300)
    plt.close()
    # plt.show()
```
