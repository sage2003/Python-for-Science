```python
r = np.linspace(0,10,100)
psi = 2*np.exp(-r) #1s
psi1 = 2 * 2**(-3/2) * (1-r/2) * np.exp(-r/2) #2s
psi2 = 1/np.sqrt(3) * 2**(-3/2) * r * np.exp(-r/2) #2p
psi3 = 2 * 3**(-3/2) * (1 - 2*r/3 + 2*r**2/27) * np.exp(-r/3) #3s
psi4 = 4*np.sqrt(2)/3 * 3**(-3/2) * r * (1-r/6) * np.exp(-r/3) #3p
psi5 = 2*np.sqrt(2)/(27*np.sqrt(5)) * 3**(-3/2) * r**2 * np.exp(-r/3) #3d
plt.plot(r,psi, label='1s')
plt.plot(r,psi1, label='2s')
plt.plot(r,psi2, label='2p')
plt.plot(r,psi3, label='3s')
plt.plot(r,psi4, label='3p')
plt.plot(r,psi5, label='3d')
plt.ylim(top=1.5)
plt.legend()
plt.ylabel('$R_{nl}$', fontsize=24)
plt.xlabel('$r$', fontsize=24)
plt.savefig('Combined Wavefunction.png')
plt.show()
```

# Radial Probability Distribution Function

```python
r = np.linspace(0, 50, 200)
psi1 = 2*np.sqrt(2)/(27*np.sqrt(5)) * 3**(-3/2) * r**2 * np.exp(-r/3) #3d
R1 = 4 * np.pi * r**2 * psi1**2 #3d
psi2 = 4*np.sqrt(2)/3 * 3**(-3/2) * r * (1-r/6) * np.exp(-r/3) #3p
R2 = 4 * np.pi * r**2 * psi2**2 #3p
psi3 = 2 * 3**(-3/2) * (1 - 2*r/3 + 2*r**2/27) * np.exp(-r/3) #3s
R3 = 4 * np.pi * r**2 * psi3**2 #3s
psi4 = 1/np.sqrt(3) * 2**(-3/2) * r * np.exp(-r/2) #2p
R4 = 4 * np.pi * r**2 * psi4**2 #2p
psi5 = 2 * 2**(-3/2) * (1-r/2) * np.exp(-r/2) #2s
R5 = 4 * np.pi * r**2 * psi5**2 #2s
psi6 = 2*np.exp(-r) #1s
R6 = 4 * np.pi * r**2 * psi6**2 #1s
plt.plot(r,R1, label='3d')
plt.plot(r,R2, label='3p')
plt.plot(r,R3, label='3s')
plt.plot(r,R4, label='2p')
plt.plot(r,R5, label='2s')
plt.plot(r,R6, label='1s')
plt.ylabel('$4\pi r^{2} R_{nl}^{2}$', fontsize=24)
plt.xlabel('$r$', fontsize=24)
plt.xlim(right=35)
plt.legend()
plt.plot()
plt.savefig('RDF.png')
plt.show()
```
